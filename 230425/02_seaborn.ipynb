```python
# 이상치의 경계 부분을 수치화
import numpy as np
```

```python
q1,q3 = np.percentile(df['total_bill'],[25,75])
print(q1,q3)
```

```python
# 상단의 경계 (q3+(iqr*1.5))
# iqr = q3 - q1
iqr = q3-q1
iqr
```

```python
upper_bound = q3 + 1.5 * iqr
upper_bound
```

```python
# 하단의 이상치 경계 q1 - (iqr*1.5)
lower_bound = q1 - 1.5 * iqr
lower_bound
```

```python
# 결측치로 바꾸는 이유는 지정을 해서 한번에 날리기 위함이다./ 이상치를 확인해서 이상치가 아닌 값들을 추출하는것. loc를 쓰면 거의 된다.
df.loc[~(df['total_bill'] > upper_bound) & ~(df['total_bill']<lower_bound)]
```

```python
def outlier(df,col):
    q1,q3 = np.percentile(df[col],[25,75])
    iqr = q3-q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr

    result = np.where((df[col] <= upper_bound) & (df[col] >= lower_bound))
    return df.loc[result]
```

```python
outlier(df,'total_bill')
```
