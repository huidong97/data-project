결측치

```python
import pandas as pd
import numpy as np
```

```python
_value = [
    [1,2,3],
    [4,np.nan,6],
    [7,8,np.nan]
]
df = pd.DataFrame(_value)
df
```

```python
#결측치를 확인
#데이터프레임의 정보 확인
df.info()
```

```python
#is()함수를 사용하여 결측치를 확인
#isna()함수 사용
df.isna().sum()
```

```python
#결측치를 제거
#dropna()는 결측치를 제거 하는 함수
#매개변수 axis : 0, 'rows'을 대입하면 해당하는 행을 삭제
#              1, 'colums'을 대입하면 해당하는 열을 삭제
df.dropna(axis = 0)
```

```python
# sales 데이터 로드
sales = pd.read_csv("../csv/Sales Records.csv")
sales.head(2)
```

```python
flg = sales['Region'] == 'Asia'
```

```python
#데이터 프레임 필터 기능
#loc[행의 조건,열의 조건]
#행의 조건을 가지고 필터
sales.loc[flg]  # ~ 연산자를 사용하면 반대로 출력이 가능함
#열의 조건을 가지고 필터
sales.loc[:,['Order Priority','Order Date']]
#행과 열의 조건을 모두 적용시켜서 필터
sales.loc[flg,['Region']]


#iloc[행의 위치, 열의 위치] loc와 다르게 index즉 위치를 집어 넣어줌
sales.iloc[1:3, 1:3]


# sales.loc[1:3]
```

```python
sales[['Region']]
```

```python
# 결측치가 포함된 데이터만 필터
df.columns = ['A', 'B', 'C']
flg_na = df.isna().any(axis=1)
```

```python
df.loc[~flg_na] #결측치가 아닌 데이터만 뽑겠다
```
